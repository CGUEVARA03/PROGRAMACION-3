```python
import pandas as pd
import matplotlib.pyplot as plt
```

Tarjetas de crédito

```python
tarjetascredito =  pd.read_csv ("/content/trx_tarjetas_credito.csv",  sep = ",")
print(tarjetascredito)
```

Uso de tarjetas por género

```python
genero = pd.crosstab(index = tarjetascredito["Gender"],  columns = "fr")
print(genero)
```

```python
genero = pd.crosstab(index = tarjetascredito ["Gender"], columns = "fr")
plt.pie(genero, labels = genero.index)
plt.xlabel("Tarjetas por género")
plt.show()
genero.to_csv("/content/genero.csv")
```

Uso de tarjetas por tipo

```python
print(tarjetascredito.groupby(by = ["Card Type"]).sum())
```

```python
tipo = pd.crosstab (index = tarjetascredito ["Card Type"], columns = "fr")
plt.pie(tipo, labels = tipo.index)
plt.xlabel("Tipo de tarjetas en uso")
plt.show()
tipo.to_csv("/content/tipo.csv")
```

Uso de tarjeta por uso


```python
print(tarjetascredito.groupby(by = ["Exp Type"]).sum().groupby(level = [0]).cumsum())
```

```python
tarjetas = pd.crosstab (index = tarjetascredito ["Exp Type"], columns = "fr")
plt.pie(tarjetas, labels = tarjetas.index)
plt.xlabel("Tipos de tarjetas usadas")
plt.show()
tarjetas.to_csv("/content/tarjetas.csv")
```

Tarjetas usadas por ciudad

```python
print(tarjetascredito.groupby(by = ["City"]).sum())
```

```python
ciudad = pd.crosstab(index = tarjetascredito ["City"], columns = "fr")
plt.pie(ciudad, labels = ciudad.index)
plt.xlabel("Tarjetas usadas en cada ciudad")
plt.show()
ciudad.to_csv("/content/ciudad.csv")
```
